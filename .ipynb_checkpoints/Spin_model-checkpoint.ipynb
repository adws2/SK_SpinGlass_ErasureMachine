```python
### Markov chain Monte Carlo simulation
### Ising model matrix calcation


import numpy as np
import matplotlib.pyplot as plt
from numpy.random import rand


class Spin_simulation():
    def __init__(self,system_size):
        self.N = system_size
        self.config = np.random.choice([-1.,1.] , size=self.N, replace = True)
        
        
    def adjacency_matrix(self, type):
        if type =='2D-Ising':
            L = int(np.sqrt(self.N))
            J = np.zeros((self.N,self.N))
            k = 0
            for a in range(self.N):
                for b in range(self.N):
                        i = a%L + ( b * L )%L**2 
                        j = (a+1)%L + ( b * L )%L**2 
                        J[i,j] = 1
                        J[j,i] = 1
                        k+=1
                        j = (a)%L + ( (b+1)*L )%L**2
                        J[i,j] = 1
                        J[j,i] = 1
                        k+=1
            
        elif type == 'SK-model':
            var = 1./np.sqrt(L/2) ## 가우시안 분산
            H = np.random.normal(0., var, size = (L,L))
            for k in range(L):  H[k,k] = 0
            J = (H + H.T)/2
            
        else:
            print('2D-Ising 또는 SK-model 을 입력하시오')
            J = 0.
        self.J = J
        
        
    def move(self,t):
        for i in range(len(self.config)):
            temp = np.zeros(self.N)
            temp[i] = self.config[i]
            dE =  2.*temp@self.J@self.config
            if dE<0.:
                self.config[i]= -self.config[i]
            elif rand()<np.exp(-dE/t):
                self.config[i]= -self.config[i]
                
                
                
                
    def MC_Sampling(self,T,sample_size, eq_T, sampling_interval):
        samples = np.zeros((self.N ,sample_size,len(T)))
        for t_idx,t in enumerate(T):
            for _ in range(eq_T): 
                Spin_simulation.move(self,t)

            for k_sample in range(sample_size):                            ### cal T
                for _ in range(sampling_interval):
                    Spin_simulation.move(self,t)
                samples[:,k_sample,t_idx] = self.config
        return samples
```
