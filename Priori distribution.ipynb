### priori distribution
- SK model : $p(J_{ij}) =$ Gaussian
- 2D-Ising : two-peaks distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand
```

```python
def hist(x, nbin=10):
    """
    :)
    """
    _min = x.min()
    _max = x.max()
    bins = np.linspace(_min,_max,nbin)

    widths = (bins[1:] + bins[:-1]) /2
    hist, e = np.histogram(x, bins, density=True)
    return widths, hist 
```

## 1. Gaussian

- Gaussian priori : L2 norm
$P_J \approx \exp{J^2}$  
$\ln P_J \approx J^2$  
$\frac{\partial \ln P_J} {\partial J} \approx J$

```python
dist = rand.normal(0, 1/np.sqrt(320),size = 10000)
a,b = hist(dist,nbin = 50)
plt.plot(a,b)
plt.show()
```

```python
## Gaussian sum is also Gaussian

def gaussian(mean = 0., std = 0.0001, size = 1000000):
    x = rand.uniform(-std*10,std*10,size)
    P_1 = 1/std/np.sqrt(2*np.pi) * np.exp((- (x - mean)**2) /(2*(std**2)) )
    P_2 = 1/std/np.sqrt(2*np.pi) * np.exp((- (x - 1)**2) /(2*(std**2)) )
    P = P_1+P_2
    P /= np.sum(P)
    samples = rand.choice(x,size = size,p = P)
    return samples
```

```python
dist = gaussian()
a,b = hist(dist,nbin = 100)
plt.plot(a,b)
plt.show()
```

```python
## addition of each Gaussian function is also Gaussian

def beta(beta = 1,ahlpa = 0.00001,size = 1000000):
    x = rand.uniform(0,1,size)
    P = np.power(x,ahlpa -1) * np.power(beta-x,ahlpa - 1)
    P /= np.sum(P)
    samples = rand.choice(x,size = size,p = P)
    return samples

dist = beta()
#a,b = hist(dist,nbin = 100)
#plt.plot(a,b)
plt.hist(dist,density = 1,bins = 100)
plt.show()
```

```python
def two_peak(size = 1000000):
    x = rand.uniform(-1,1.5,size)
    P = np.exp(-(x**2)*((x-1)**2)/0.001)
    P /= np.sum(P)
    samples = rand.choice(x,size = size,p = P)
    return samples

dist = two_peak()
#a,b = hist(dist,nbin = 100)
#plt.plot(a,b)
plt.hist(dist,density = 1,bins = 100)
plt.show()
```
